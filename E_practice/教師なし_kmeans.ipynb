# kmeans

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
test_data = np.random.rand(1000,2)
```

```python
def init_centroid(data,k,n_data):
    idx = np.random.permutation(n_data)[:k]
    centroids = data[idx]
    
    return centroids 
def compute_distence(data,k,n_data,centroids):
    distances = np.zeros((n_data,k))
    for centroid_idx in range(k):
        dist = np.sqrt(np.sum((data - centroids[centroid_idx])**2,axis = 1))
        distances[:,centroid_idx] = dist
    return distances

def show_scatter(data,clusters):
    color = ['r','g','b']
    for k_idx in range(k):
        d = data[clusters == k_idx]
        plt.scatter(d[:,0],d[:,1],c = color[k_idx])
```

```python
data = test_data
k = 3
n_data = test_data.shape[0]
iter = 300
n_data , n_features = data.shape
###ランダムで中心選ぶ
centroids  = init_centroid(data,k,n_data)
new_clusters = np.zeros(n_data)
clusters = np.zeros(n_data)
for i in range(iter):
    show_scatter(data,clusters)
    plt.show()
    ###距離を計算
    distances = compute_distence(data,k,n_data,centroids)
    ###各データに、セントロイドから距離最短のクラスタを付与
    new_clusters = np.argmin(distances,axis = 1)
    ###セントロイドクラスタごとにセントロイドを再計算
    for idx_centroids in range(k):
    centroids[idx_centroids] = data[new_clusters == idx_centroids].mean(axis = 0)
    ### 収束した場合は中断
    if (new_clusters == clusters).all():
    break
    clusters = new_clusters
```
